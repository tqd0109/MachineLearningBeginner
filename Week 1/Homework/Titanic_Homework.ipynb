### Analyze Titatic dataset

```python
# Importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.sample(5)
```

```python
data.describe(include='all')
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
print(data.isnull().sum())
```

```python
# Create new columns for family size
data['FamilySize'] = data ['SibSp'] + data['Parch'] + 1
data['IsAlone'] = 1 #initialize to yes/1 is alone
data['IsAlone'].loc[data['FamilySize'] > 1] = 0 # now update to no/0 if family size is greater than 1
data.sample(5)
```

```python
sns.countplot(x='IsAlone', hue='Survived', data=data)
```

```python
sns.countplot(x='Pclass', hue='Survived', data=data)
# People in class 1 and 2 have more probability to live than class 3
```

```python
sns.countplot(x='Sex', hue='Survived', data=data)
# If passenger was a man, he probably died
```

```python
plt.hist(x='Age', data=data[data['Survived']==0], color='blue', alpha=0.5)
plt.hist(x='Age', data=data[data['Survived']==1], color='red', alpha=0.5)
# Children had more chance to live
# Almost all of the people that older than 65 died.
```

```python
sns.barplot(x = 'FamilySize', y = 'Survived', data=data)
```
